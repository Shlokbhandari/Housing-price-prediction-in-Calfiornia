```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
df = pd.read_csv("housing.csv")
```

```python
df['income_cat'] = pd.cut(df['median_income'], bins = [0, 1.5, 3.0, 4.5, 6.0, 7.5, np.inf], labels = [1, 2, 3 ,4 ,5, 6 ])
```

```python
split = StratifiedShuffleSplit(n_splits = 1, test_size = 0.2, random_state = 42)
for train_index, test_index in split.split(df, df['income_cat']):
    strat_train_set = df.loc[train_index]
    strat_test_set = df.loc[test_index]
```

```python
#Remove income_cat column
for x in (strat_train_set, strat_test_set, df):
    x.drop("income_cat", axis = 1, inplace = True)
```

```python
df = strat_train_set.copy()
```

```python
housing = strat_train_set.drop("median_house_value", axis = 1)
housing_labels = strat_train_set["median_house_value"].copy
```

```python
housing = housing.drop("ocean_proximity", axis = 1)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
```

```python
mypipeline = Pipeline([
             ("imputer", SimpleImputer(strategy="median")),
             ("Scaler", StandardScaler()),
              ])
```

```python
housing_pip = mypipeline.fit_transform(housing)
```

```python
housing = pd.DataFrame(housing_pip, columns = housing.columns, index = housing.index)
```
