```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("housing.csv")
```

```python
df['income_cat'] = pd.cut(df['median_income'], bins = [0, 1.5, 3.0, 4.5, 6.0, 7.5, np.inf], labels = [1, 2, 3 ,4 ,5, 6 ])
```

```python
df['income_cat'].value_counts().sort_index().plot.bar(rot=0, grid = True)
plt.title("Income Categories Distribution")
plt.xlabel("Income median")
plt.ylabel("Number of Instances")
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
split = StratifiedShuffleSplit(n_splits = 1, test_size = 0.2, random_state = 42)
for test_index, train_index in split.split(df, df['income_cat']):
    strat_train_set = df.loc[train_index]
    strat_test_set = df.loc[test_index]
```

```python
strat_test_set['income_cat'].value_counts().sort_index().plot.bar(rot=0, grid = True)
plt.title("Income Categories Distribution")
plt.xlabel("Income median")
plt.ylabel("Number of Instances")
#The bar plot of Train set is same as the main data
```

```python
strat_test_set['income_cat'].value_counts().sort_index().plot.bar(rot=0, grid = True)
plt.title("Income Categories Distribution")
plt.xlabel("Income median")
plt.ylabel("Number of Instances")
#The bar plot of Test set is same 
```
